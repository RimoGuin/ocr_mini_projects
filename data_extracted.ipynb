```python
import easyocr
import numpy as np

# Initialize the reader
reader = easyocr.Reader(['en'], gpu=True)

# Load images - Ensure this is an array of images, not paths
cv2_images = np.array(image_path)  # Replace image_path with actual image array

# Parameters for repetition
img_repeat, num_loop = 5, 1

# Repeat the images
cv2_images_repeat1 = np.repeat(cv2_images, repeats=img_repeat, axis=0)
cv2_images_repeat2 = cv2_images_repeat1.copy()

# Initialize result dictionary
result = {}

# Loop through images and perform OCR
for i, img in enumerate(cv2_images):
    print(img)
    result[i] = reader.readtext(img)

# Batch reading for multiple loops
#reader = easyocr.Reader(['en'], cudnn_benchmark=True)
#for i in range(num_loop):
#    reader.readtext_batched(cv2_images_repeat2)
```

```python
reader = easyocr.Reader(['en'], cudnn_benchmark=True)
cv2_images = image_path

img_repeat, num_loop = 5, 1
cv2_images_repeat1 = np.repeat(cv2_images, repeats=img_repeat, axis=0)
cv2_images_repeat2 = cv2_images_repeat1.copy()

i = 0
for img in cv2_images:
    print(img)
    result[i] = reader.readtext(img)
    i = i + 1

reader = easyocr.Reader(['en'], cudnn_benchmark=True)
for i in range(num_loop):
    reader.readtext_batched(cv2_images_repeat2)
```
